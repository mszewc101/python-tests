# Python od podstaw 000 - prework instalacja pythona, IDE

# 1. Instalacja Pythona

## **Windows**

### 1. Pobranie instalatora
- Wejdź na oficjalną stronę Pythona: [https://www.python.org](https://www.python.org).
- Przejdź do sekcji **Downloads** i kliknij na wersję odpowiednią dla Windows (np. "Download Python 3.x.x").

### 2. Uruchomienie instalatora
- Otwórz pobrany plik `.exe`.
- **Zaznacz opcję**: **"Add Python to PATH"** (bardzo ważne, aby Python działał w wierszu poleceń).
- Wybierz opcję **Install Now** lub **Customize installation**, jeśli chcesz określić lokalizację instalacji.

### 3. Sprawdzenie instalacji
- Otwórz **Wiersz poleceń (cmd)**: wciśnij **Windows + R**, wpisz `cmd` i naciśnij Enter.
- Wpisz polecenie:
  ```bash
  python --version
  ```
- Powinieneś/aś zobaczyć zainstalowaną wersję Pythona.

## **macOS**

### 1. Sprawdzenie, czy Python jest zainstalowany
- Na większości komputerów z macOS Python 2.x jest preinstalowany, ale lepiej zainstalować nową wersję.
- Otwórz **Terminal** i wpisz:
  ```bash
  python3 --version
  ```

### 2. Pobranie instalatora
- Przejdź na stronę [https://www.python.org](https://www.python.org).
- W sekcji **Downloads** wybierz instalator dla macOS i pobierz go.

### 3. Uruchomienie instalatora
- Otwórz pobrany plik instalacyjny `.pkg`.
- Postępuj zgodnie z instrukcjami wyświetlanymi na ekranie, aby zainstalować Pythona.

### 4. Sprawdzenie instalacji
- Po zakończeniu instalacji upewnij się, że Python został poprawnie zainstalowany.
- Otwórz Terminal (w Finderze wyszukaj "Terminal").
  ```bash
  python3 --version
  ```
- Powinieneś/aś zobaczyć zainstalowaną wersję Pythona.

### 5. Opcjonalnie: Instalacja za pomocą Homebrew
- Jeśli korzystasz z Homebrew (menedżera pakietów dla macOS), możesz zainstalować Pythona poleceniem:
```bash
brew install python
```


# 2. IDE - Integrated Development Environment
### Najpopularniejsze IDE, Pycharm(community - darmowy) lub VS Code. Wybierzcie które bardziej wam odpowiada(jeden wystarczy). Szkolenie będę prowadził na VS Code, ale zaprezentuje też użycie Pycharma.

## PyCharm Community - Windows:
### 1. Pobranie instalatora
1. Przejdź na stronę [PyCharm](https://www.jetbrains.com/pycharm/).
2. Kliknij przycisk **Download**.
3. Wybierz wersję **Community** (darmową) i pobierz instalator.

### 2. Uruchomienie instalatora
1. Otwórz pobrany plik `.exe`.
2. Postępuj zgodnie z instrukcjami wyświetlanymi w kreatorze instalacji:
   - Wybierz lokalizację instalacji (domyślnie jest ustawiona automatycznie).
   - Na ekranie z opcjami zaznacz:
     - **Create Desktop Shortcut** – jeśli chcesz mieć skrót na pulpicie.
     - **Update PATH Variable** – aby dodać PyCharm do zmiennych systemowych (opcjonalne, ale pomocne).
3. Kliknij **Install**, aby rozpocząć proces instalacji.

### 3. Pierwsze uruchomienie
1. Po zakończeniu instalacji uruchom PyCharm:
   - Kliknij ikonę PyCharm na pulpicie lub znajdź ją w menu Start.
2. Przy pierwszym uruchomieniu:
   - Możesz zaimportować ustawienia z innej instalacji lub pominąć ten krok.
   - Zostaniesz poproszony o wybranie motywu interfejsu (jasny lub ciemny).
   - Możesz skonfigurować dodatkowe wtyczki, jeśli są potrzebne.

### 4. Utworzenie nowego projektu
1. Po uruchomieniu PyCharm kliknij **New Project**.
2. Wybierz lokalizację projektu oraz interpreter Pythona (PyCharm automatycznie wyszuka zainstalowanego Pythona, jeśli jest dostępny).

## PyCharm Community - macOS

### 1. Pobranie instalatora
1. Przejdź na stronę [PyCharm](https://www.jetbrains.com/pycharm/).
2. Kliknij przycisk **Download**.
3. Wybierz wersję **Community** (darmową) i pobierz instalator w formacie `.dmg`.

### 2. Instalacja PyCharm
1. Otwórz pobrany plik `.dmg`.
2. Przeciągnij ikonę PyCharm do folderu **Applications**.
3. Poczekaj, aż pliki zostaną skopiowane do systemu.

### 3. Uruchomienie PyCharm
1. Otwórz Finder, przejdź do folderu **Applications** i kliknij dwukrotnie ikonę PyCharm.
2. Przy pierwszym uruchomieniu:
   - Jeśli pojawi się komunikat o aplikacji pobranej z internetu, kliknij **Otwórz**.
   - Możesz zaimportować ustawienia z wcześniejszej wersji lub pominąć ten krok.
   - Wybierz motyw interfejsu (jasny lub ciemny).
   - Skonfiguruj dodatkowe wtyczki, jeśli chcesz.

### 4. Utworzenie nowego projektu
1. Po uruchomieniu PyCharm kliknij **New Project**.
2. Wybierz lokalizację projektu oraz interpreter Pythona (PyCharm automatycznie wyszuka zainstalowanego Pythona, jeśli jest dostępny).



## Visual Studio Code - Windows

### 1. Pobranie instalatora
1. Przejdź na oficjalną stronę Visual Studio Code: [https://code.visualstudio.com/](https://code.visualstudio.com/).
2. Kliknij przycisk **Download for Windows**, aby pobrać instalator.

### 2. Uruchomienie instalatora
1. Otwórz pobrany plik instalacyjny `.exe`.
2. Postępuj zgodnie z instrukcjami wyświetlanymi w kreatorze instalacji:
   - **Zaakceptuj licencję użytkownika**.
   - Wybierz lokalizację instalacji (domyślna jest zazwyczaj odpowiednia).
   - Zaznacz opcje dodatkowe:
     - **Add to PATH** – umożliwia uruchamianie VS Code z poziomu wiersza poleceń.
     - **Register Code as default editor for supported file types** – ustawia VS Code jako domyślny edytor plików.
3. Kliknij **Install**, aby rozpocząć instalację.

### 3. Uruchomienie VS Code
1. Po zakończeniu instalacji możesz uruchomić Visual Studio Code bezpośrednio z kreatora instalacji, klikając **Finish**.
2. Jeśli zamknąłeś kreator, znajdź ikonę VS Code na pulpicie lub w menu Start i uruchom aplikację.

### 4. Konfiguracja środowiska
1. Po pierwszym uruchomieniu możesz zainstalować wtyczki, takie jak:
   - **Python** – do pracy z językiem Python.

## Visual Studio Code - macOS

### 1. Pobranie instalatora
1. Przejdź na oficjalną stronę Visual Studio Code: [https://code.visualstudio.com/](https://code.visualstudio.com/).
2. Kliknij przycisk **Download for macOS**, aby pobrać instalator w formacie `.zip`.

### 2. Instalacja Visual Studio Code
1. Otwórz pobrany plik `.zip`. System automatycznie rozpakowuje plik, tworząc aplikację **Visual Studio Code**.
2. Przeciągnij aplikację **Visual Studio Code** do folderu **Applications**, aby ją zainstalować.

### 3. Uruchomienie VS Code
1. Otwórz Finder, przejdź do folderu **Applications** i kliknij dwukrotnie ikonę **Visual Studio Code**.
2. Jeśli pojawi się komunikat o aplikacji pobranej z internetu, kliknij **Otwórz**.

### 4. Konfiguracja środowiska
1. Po pierwszym uruchomieniu możesz dostosować VS Code do swoich potrzeb, instalując wtyczki:
   - Kliknij ikonę **Extensions** (symbol czterech kwadratów w lewym pasku).
   - Wyszukaj potrzebne rozszerzenia, takie jak:
     - **Python** – do pracy z językiem Python.
