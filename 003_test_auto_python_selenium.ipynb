# Czym Jest Selenium:
- Selenium to zestaw narzędzi i bibliotek, które pozwalają na automatyzację
działań wykonywanych w przeglądarkach internetowych. Może być używany
do testowania aplikacji webowych, skryptowania działań użytkownika i wykonywania
różnych zadań związanych z przeglądarką.



# Zalety `selenium`
1. **Wsparcie dla wielu przeglądarek:** Selenium obsługuje wiele popularnych przeglądarek internetowych, w tym Google Chrome, Mozilla Firefox, Microsoft Edge, Safari, Opera, oraz inne. Dzięki temu możesz pisać testy lub skrypty, które działają na różnych przeglądarkach.

2. **Wsparcie dla wielu języków programowania:** Selenium jest dostępne w wielu językach programowania, w tym Python, Java, C#, Ruby, i wiele innych.

3. **Automatyzacja interakcji z przeglądarką:** Selenium pozwala symulować interakcje użytkownika z przeglądarką, takie jak klikanie, wypełnianie formularzy, nawigacja po stronach, i wiele innych. Możesz pisać skrypty, które testują funkcjonalność aplikacji internetowych.

4. **Testowanie automatyczne:** Selenium jest często używane do automatyzacji testów jednostkowych, testów akceptacyjnych (end-to-end), i testów wydajnościowych. Pozwala na powtarzalne i skalowalne testowanie aplikacji.

5. **Integracja z różnymi narzędziami:** Selenium można integrować z różnymi narzędziami i frameworkami testowymi, takimi jak TestNG, JUnit, Robot Framework, czy PyTest. Dzięki temu można dostosować proces testowania do własnych potrzeb.

6. **Zaawansowane scenariusze testowe:** Selenium pozwala na tworzenie zaawansowanych scenariuszy testowych, takich jak testy wielokrokowe, testy wielu przeglądarek, testy równoległe, i wiele innych.

7. **Community i dokumentacja:** Selenium ma silne wsparcie społeczności. Istnieje wiele dostępnych materiałów edukacyjnych, kursów online, oraz dokumentacja, która pomaga w nauce i rozwiązywaniu problemów związanych z Selenium.


# Wady `selenium`

1. **Konieczność dostępu do przeglądarek:** Selenium wymaga dostępu do
przeglądarek internetowych, co może być problematyczne na niektórych środowiskach.

2. **Czułość na zmiany w interfejsie użytkownika:** Testy napisane w Selenium
mogą przestawać działać, jeśli strona internetowa często zmienia swoją strukturę HTML.

3. **Wolniejsza wydajność:** Selenium działa poprzez interakcję z przeglądarką,
co sprawia, że jest wolniejsze niż inne rodzaje testów.

4. **Brak wsparcia dla wielowątkowości:** Selenium nie oferuje natywnej obsługi
wielowątkowości, co może być problemem w przypadku próby uruchamiania testów równolegle.

5. **Zabezpieczenia przeglądarek:** Niektóre strony internetowe wprowadzają
zabezpieczenia, które utrudniają automatyzację za pomocą Selenium.

6. **Koszty utrzymania:** Utrzymanie testów w Selenium może wymagać dużo pracy,
szczególnie w przypadku aplikacji, które często się zmieniają.


# Instalacja selenium dla języka `python`

1. **Sprawdź, czy masz zainstalowanego Pythona:** Upewnij się, że masz zainstalowanego Pythona na swoim komputerze. Selenium jest kompatybilne z Pythonem 3.5 i nowszymi wersjami.

3. **Zainstaluj Selenium za pomocą pip:**

   Aby zainstalować Selenium, wykonaj polecenie

   ```bash
   pip install selenium
   ```

4. **Pobierz przeglądarkę WebDriver:**

   Selenium wymaga WebDrivera, który jest odpowiedzialny za kontrolowanie
   przeglądarki. Musisz pobrać odpowiednią wersję WebDrivera dla przeglądarki,
   którą planujesz używać z Selenium. Na przykład, jeśli chcesz używać
   Google Chrome, pobierz ChromeDriver. Po pobraniu ChromeDrivera,
   umieść go w odpowiednim miejscu na swoim komputerze (np. w katalogu,
   który jest w zmiennej środowiskowej PATH).

5. **Sprawdź instalację:**

   Aby upewnić się, że Selenium zostało pomyślnie zainstalowane, uruchom
   interpreter Pythona w terminalu i wpisz

   `from selenium import webdriver`.

   Jeśli nie otrzymasz żadnych błędów, oznacza to, że Selenium zostało
   pomyślnie zainstalowane.


## Ad 3. Instrukcja Instalacji WebDriver dla Google Chrome

1. **Sprawdź wersję Google Chrome:**

   Upewnij się, że masz zainstalowaną przeglądarkę Google Chrome i sprawdź jej
   wersję. Aby to zrobić, otwórz Google Chrome i kliknij na trzy kropki w prawym
   górnym rogu, a następnie przejdź do "Pomoc" > "Informacje o Google Chrome".
   Znajdziesz tam informacje o wersji przeglądarki.

2. **Pobierz odpowiedni ChromeDriver:**

   Następnie musisz pobrać odpowiednią wersję ChromeDriver, która jest
   kompatybilna z Twoją wersją przeglądarki i systemem operacyjnym. Otwórz stronę
   pobierania ChromeDrivera: [ChromeDriver Downloads](https://sites.google.com/chromium.org/driver/)

   Wybierz wersję ChromeDrivera, która odpowiada Twojej wersji przeglądarki i
   sytemu operacyjnego. Pobierz plik ZIP z ChromeDriverem.

3. **Rozpakuj ChromeDriver:**

   Po pobraniu ChromeDrivera, rozpakuj plik ZIP do wybranego katalogu na swoim komputerze.

4. **Dodaj ścieżkę do ChromeDriver do zmiennej PATH:**

   Aby Selenium mogło znaleźć ChromeDriver, musisz dodać ścieżkę do katalogu,
   w którym znajduje się ChromeDriver, do zmiennej środowiskowej PATH na swoim
   komputerze. Dzięki temu Selenium będzie w stanie automatycznie go odnaleźć.

   - Jeśli korzystasz z systemu Windows, dodaj ścieżkę do ChromeDriver do zmiennej
   PATH w ustawieniach systemowych.
      1. **Otwórz ustawienia systemowe:** Kliknij prawym przyciskiem myszy na
      ikonie "Mój komputer" (lub "Komputer" w zależności od wersji systemu Windows) na pulpicie i wybierz "Właściwości".

      3. **Otwórz zaawansowane ustawienia systemu:** W lewym panelu wybierz "Zaawansowane ustawienia systemu".

      4. **Otwórz zmienne środowiskowe:** W sekcji "Zaawansowane" na karcie "Ogólne" kliknij przycisk "Zmienne środowiskowe".

      5. **Edytuj zmienne środowiskowe użytkownika:** W sekcji "Zmienne użytkownika"
      znajdź zmienną o nazwie "Path" lub "PATH" i wybierz ją. Następnie kliknij przycisk "Edytuj".

      6. **Dodaj ścieżkę do WebDriver:** Kliknij przycisk "Nowy" i wprowadź
      ścieżkę do katalogu, w którym znajduje się WebDriver (np. ścieżka do
      katalogu z ChromeDriverem). Upewnij się, że ścieżka jest poprawna i nie zawiera żadnych błędów.

      7. **Zapisz zmiany:** Po dodaniu ścieżki do WebDrivera, zatwierdź
      wszystkie otwarte okna klikając "OK" lub "Zastosuj". Wartość zmiennej
      PATH zostanie zaktualizowana.

      8. **Sprawdź instalację:** Otwórz nowe okno wiersza poleceń (Command Prompt)
      i wpisz komendę, która uruchamia WebDriver (np. "chromedriver" dla ChromeDrivera).
      Jeśli WebDriver jest dostępny w zmiennej PATH, nie powinieneś otrzymać
      błędu i WebDriver powinien się uruchomić.


   - Jeśli korzystasz z systemu Unix/Linux (np. macOS lub Linux), możesz dodać
   ścieżkę do ChromeDriver do zmiennej PATH, wykonując polecenie np.:

     ```bash
     export PATH=$PATH:/ścieżka/do/katalogu/z/ChromeDriverem
     ```

     Upewnij się, że zastąpisz `/ścieżka/do/katalogu/z/ChromeDriverem` właściwą
     ścieżką do katalogu, w którym znajduje się ChromeDriver.

## Test możemy poprawić do bardziej modularnego i skalowalnego używjąc funkcji

```python
from selenium import webdriver
from time import sleep

def get_website_title(url):
    driver = webdriver.Safari()
    driver.get(url)
    title = driver.title
    driver.quit()
    return title

def test_codebrainers_title():
    url = "https://codebrainers.pl/"
    expected_title = "CodeBrainers – Prawdopodobnie najlepszy bootcamp w Polsce!"
    website_title = get_website_title(url)
    assert expected_title == website_title

test_codebrainers_title()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import *

driver = webdriver.Safari()
driver.get("https://google.com")
print(driver.title)
driver.implicitly_wait(3)
accept_all = driver.find_element(By.ID, "L2AGLb")
accept_all.click()

driver.implicitly_wait(3)
search = driver.find_element(By.ID, "APjFqb")
search.send_keys("selenium test")

driver.implicitly_wait(3)
search.send_keys(Keys.RETURN)

sleep(4)

results = driver.find_element(By.ID, "result-stats")
print(results.text)

assert results.text is not None
driver.quit()


```
